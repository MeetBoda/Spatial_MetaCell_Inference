### Approach 4 code

```python
import scanpy as sc
sc_data_covet= sc.read_h5ad('/kaggle/input/cs690-spatial-metacell-inference-merfish-data/sc_data_with_covet_results.h5ad')
sc_data_covet.obsm['COVET_SQRT'].shape
```

```python
C = sc_data_covet.obsm["COVET_SQRT"]    # shape: (n_cells, g, g)
print(C.shape)
```

```python
if C.ndim == 2:
    n, d = C.shape
    g = int(np.sqrt(d))
    C = C.reshape(n, g, g)
```

```python
import numpy as np
from sklearn.metrics import pairwise_distances

# Flatten each COVET_SQRT matrix into a vector
C_flat = C.reshape(C.shape[0], -1)

# AOT distance = Euclidean distance between flattened COVET_SQRT matrices
AOT_dist = pairwise_distances(C_flat, metric="euclidean")
```

```python
AOT_dist.shape
```

```python
from sklearn.neighbors import NearestNeighbors

k = 20   # number of neighbors; tune based on density
nbrs = NearestNeighbors(n_neighbors=k, metric="euclidean")
nbrs.fit(C_flat)
distances, indices = nbrs.kneighbors(C_flat)
```

```python
import scanpy as sc
import scipy.sparse as sp

n_cells = C_flat.shape[0]

# Build adjacency = mutual kNN graph
row = np.repeat(np.arange(n_cells), k)
col = indices.flatten()
data = np.ones(len(row))

adj = sp.csr_matrix((data, (row, col)), shape=(n_cells, n_cells))
adj = adj.maximum(adj.T)  # make symmetric
```

```python
!pip install leidenalg
```

```python
import numpy as np
import igraph as ig
import leidenalg

# Build graph from adjacency matrix
coo = adj.tocoo()
edges = list(zip(coo.row, coo.col))

g = ig.Graph(edges=edges, directed=False)
g.simplify()

partition = leidenalg.find_partition(
    g, leidenalg.RBConfigurationVertexPartition
)

niche_clusters = np.array(partition.membership)
```

```python
sc_data_covet.obs["niche_id"] = niche_clusters.astype(str)
```

```python
import anndata as ad
import numpy as np
import os
from MetaQ_sc import run_metaq

adata = sc_data_covet.copy()

# Ensure niche_id exists
assert "niche_id" in adata.obs, "Run niche clustering first!"

# Get list of niches
niches = sorted(adata.obs["niche_id"].unique().tolist(),reverse=True,key= lambda x:int(x))

print("Found niches:", niches)

# Total metacells you want
M_total = 500

# Total cells
N_total = adata.n_obs

# Directory to store all niche-specific h5ad files
os.makedirs("niche_metaq_inputs", exist_ok=True)

# Keep results directory
os.makedirs("niche_metaq_outputs", exist_ok=True)

for niche in niches:
    print(f"\n=== Processing niche {niche} ===")

    # Subset to that niche
    ad_n = adata[adata.obs["niche_id"] == niche].copy()
    N_c = ad_n.n_obs

    print(f"Cells in niche {niche}: {N_c}")

    # Compute metacells for this niche
    M_c = max(1, int(np.ceil((N_c / N_total) * M_total)))
    print(f"Metacells allocated: {M_c}")

    # Save subset
    niche_h5ad = f"niche_metaq_inputs/niche_{niche}.h5ad"
    ad_n.write_h5ad(niche_h5ad, compression="gzip")

    # Run MetaQ
    try:
        print(f"Running MetaQ for niche {niche}...")
        run_metaq(
            data_path=[niche_h5ad],
            data_type=["RNA"],
            metacell_num=M_c,
            save_name=f"niche_{niche}_metaQ",
        )
        print(f"✔ Completed niche {niche}")

    except Exception as e:
        print(f"✘ ERROR in niche {niche}: {str(e)}")
        # Continue to next niche
        continue
```

```python
import anndata as ad
import glob
import numpy as np

mc_files = sorted(glob.glob("niche_*_metaQ_metacell.h5ad"))
assign_files = sorted(glob.glob("niche_*_metaQ_cell2metacell.h5ad"))

mc_list = []
assign_list = []

for mc_f, asg_f in zip(mc_files, assign_files):
    niche_id = mc_f.split("_")[1]   # extract niche number
    
    mc = ad.read_h5ad(mc_f)
    asg = ad.read_h5ad(asg_f)

    # tag metacells with niche id
    mc.obs["niche_id"] = niche_id
    asg.obs["niche_id"] = niche_id

    # rename metacell IDs to avoid overlap between niches
    mc.obs_names = [f"{niche_id}_{x}" for x in mc.obs_names]
    asg.obs["metacell"] = [f"{niche_id}_{x}" for x in asg.obs["metacell"]]

    mc_list.append(mc)
    assign_list.append(asg)


adata_mc = ad.concat(mc_list, axis=0, join="outer")
adata_assign = ad.concat(assign_list, axis=0, join="outer")

adata_mc.write_h5ad("MetaQ_spatial_all_metacells.h5ad")
adata_assign.write_h5ad("MetaQ_spatial_cell_assignments.h5ad")

print("Saved:")
print("  → MetaQ_spatial_all_metacells.h5ad")
print("  → MetaQ_spatial_cell_assignments.h5ad")
```
